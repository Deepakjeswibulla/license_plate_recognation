```python
import subprocess
import sys
import importlib

import importlib.util

spec = importlib.util.find_spec("numpy")
if spec is not None:
    print("Uninstalling existing numpy...")
    subprocess.run([sys.executable, "-m", "pip", "uninstall", "-y", "numpy"], check=False)
else:
    print("No existing numpy found.")
```

```python
from ultralytics import YOLO
import shutil
import os

DATASET_DIR = "License-Plate-Detection-1"  # âš ï¸ CHANGE IF VERSION DIFFERENT
DATA_YAML = os.path.join(DATASET_DIR, "data.yaml")

def train():
    model = YOLO("yolov8n.pt")  # Pretrained backbone

    model.train(
        data=DATA_YAML,
        epochs=60,
        imgsz=640,
        batch=8,
        workers=4,
        project="runs",
        name="license_plate_detector"
    )

def export_model():
    src = "runs/detect/license_plate_detector/weights/best.pt"
    dst = "license_plate.pt"

    if os.path.exists(src):
        shutil.copy(src, dst)
        print("âœ… Model saved as license_plate.pt")
    else:
        print("âŒ best.pt not found")

train()
export_model()
```
